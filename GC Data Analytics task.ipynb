<h2>Linear Regression from scratch</h2>

```python
class FromScratchLR:

    def __init__(self):
        self.coef_ = None
        self.intercept_= None

    def fit(self, X_train, y_train):
        X_train= np.insert(X_train,0,1,axis=1)
        wt = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
        self.intercept_ = wt[0]
        self.coef_= wt[1:]

    def predict(self,X_test):
        y_pred = np.dot(X_test,self.coef_)+ self.intercept_
        return y_pred
```

<b><i>Linear Regression is implemented from scratch using the Normal Equation:
=()^1()
A column of ones was added to the input feature matrix to learn the intercept term. Model coefficients and intercept were computed using NumPy linear algebra operations.<b/><i/>

```python
lr2 = FromScratchLR()
lr2.fit(X_train, y_train)
```

```python
y_test_pred2 = lr2.predict(X_test)
y_val_pred2 = lr2.predict(X_val)
```

```python
print("From scratch LR R2 score : ",r2_score(y_val,y_val_pred2))
```

<b><i>The R2 score from scratch model closely matches the R2 score from scikit learn model <b/><i/>
